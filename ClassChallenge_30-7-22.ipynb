# 1 .Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave

```python
pf_data['Date'] = pf_data['Date'].apply(pd.to_datetime)
pf_data['week'] = pf_data['Date'].dt.isocalendar().week
pf_data
```

```python
weekly_data = pf_data.groupby(['Agent Name', 'week']).mean('Average Rating').reset_index()
weekly_data
```
